```python
df=nfl.import_pbp_data([2019,2020,2021], downcast=True, cache=False, alt_path=None)
```

```python
df=df.reset_index(drop=True)
```

```python
#explore
qb_feats=['season','passer_id','passer','pass','complete_pass','interception','sack','yards_gained','touchdown']

groupby_feats=['season','passer_id','passer']

qb_df=(df.loc[:,qb_feats].groupby(groupby_feats,as_index=False).sum())

qb_df.sample(10)
qb_df.head()
```

```python
for y in ['yards_gained','complete_pass','pass','interception','sack']:
    sns.regplot(data=qb_df,x='touchdown',y=y)
    plt.title(f'touchdowns and {y}')
    plt.show()
```

```python
_df = qb_df.copy()

# adding 1 to season for past and future differentiation
_df['season'] = _df['season'].add(1)

new_qb_df = (qb_df
             .merge(_df, 
                    on=['season', 'passer_id', 'passer'], 
                    suffixes=('', '_prev'), 
                    how='left'))

new_qb_df.sample(10)
```

```python
for y in ['touchdown_prev','yards_gained_prev', 
          'complete_pass_prev', 'pass_prev', 
          'interception_prev','sack_prev']:
  sns.regplot(data=new_qb_df, x='touchdown', y=y)
  plt.title(f"touchdowns and {y}")
  plt.show()
```

```python
#Machine learning


from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error
from scipy.stats import pearsonr
```

```python
new_qb_df.sample(10)
```

```python
features = ['pass_prev', 'complete_pass_prev', 
            'interception_prev', 'sack_prev',
            'yards_gained_prev', 'touchdown_prev']

# predicting "current" season touchdowns 
target = 'touchdown' 

#subset of data that has no null values
model_data = (new_qb_df
              .dropna(subset=features+[target]))

# train on 2020 data  
train_data = (model_data
              .loc[model_data['season']==2020])

# test on 2021 data (data the model hasn't seen)
test_data = (model_data
             .loc[model_data['season']==2021])

# initializing the linear regression
model = LinearRegression()

# fitting, or "training" the model on the training data
model.fit(train_data.loc[:, features], 
          train_data[target])

# predicting on the test data 
preds = model.predict(test_data.loc[:, features])

# setting an index so your predictions match the correct rows
preds = pd.Series(preds, index=test_data.index)

# joining predictions back to your test dataset
test_data['preds'] = preds
```

```python

rmse = mean_squared_error(test_data['touchdown'], test_data['preds'])**0.5
r2 = pearsonr(test_data['touchdown'], test_data['preds'])[0]**2
print(f"rmse: {rmse}\nr2: {r2}")
```

```python
sns.regplot(data=test_data, x='touchdown', y='preds')
plt.title('touchdown and predictions')
plt.show()
```
