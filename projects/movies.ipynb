```python
#importing libraries
import pandas as pd 
import seaborn as sns
import numpy as np  
import matplotlib
import matplotlib.pyplot as plt 
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize']=(12,8)

#importing the data
df= pd.read_csv('/Users/shravann/Downloads/cyclistic/Portfolio db/movies.csv')
```

```python
#checking out the imported data
df.head()
```

```python
#checking for nulls and whatnot
for col in df.columns:
    pct_missing= np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
#checking out data types of the dataset
df.dtypes
```

```python
#changing table view settings 

#pd.set_option('display.max_rows',None)
```

```python
df.sort_values(by=['gross'],inplace=False,ascending=False)
```

```python
#removing duplicates
df.drop_duplicates()
```

```python
#Scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross earnings')
plt.xlabel('Gross Earnings')

plt.ylabel('Budget for film')
plt.show()
```

```python
df.head(5)
```

```python
#plot budget vs gross using seaborn

sns.regplot(x='budget',y='gross',data=df,scatter_kws={"color":"blue"},line_kws={"color":"red"})
```

```python
#correlation analysis, an easy and productive way to EDA any dataset
df.corr(numeric_only=True)

#df.corr is not numeric only by default, the method is pearson by default
```

```python
#pearson, kendall, spearman, different methods of correlation
df.corr(method='pearson',numeric_only=True)

```

```python
df.corr(method ='spearman',numeric_only=True)
```

```python
#budget and gross are highly correlated, so are gross and votes...it also seems runtime of a movie seems to matter to some extent

#so here's a heatmap to show it

correlation_matrix=df.corr(method='spearman',numeric_only=True)

sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation between various features of a movie')
plt.xlabel('Movie Features')
plt.ylabel('Movie features')
plt.show()
```

```python
correlation_matrix=df.corr(method='pearson',numeric_only=True)

sns.heatmap(correlation_matrix,annot=True)
plt.title('Correlation between various features of a movie')
plt.xlabel('Movie Features')
plt.ylabel('Movie features')
plt.show()
```
