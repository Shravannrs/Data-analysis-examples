```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
df_tracks = pd.read_csv("Users/shravann/Downloads/archive/tracks.csv")
df_tracks = df_tracks.replace("Users/shravann/Downloads/archive/tracks.csv", np.nan, regex=True)
df_tracks.head()
```

```python
most_popular = df_tracks.query('popularity>90', inplace = False).sort_values('popularity',ascending= False)
most_popular[:10]
```

```python
df_tracks.set_index("release_date",inplace=True)
```

```python

df_tracks.index=pd.to_datetime(df_tracks.index,format='ISO8601')
df_tracks.head()
```

```python
df_tracks[["artists"]].iloc[18]
```

```python
df_tracks.head()
```

```python
df_tracks["duration"]=df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms", inplace=True, axis=1)
```

```python
corr_df = df_tracks.drop(["key", "explicit","id","id_artists", "mode","name","artists"], axis=1).corr(method="pearson")
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt=".1g",vmin=-1,vmax=1,center=0,cmap="inferno",linewidth=1,linecolor='blue')
heatmap.set_title("Correlation Heatmap")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates,format='ISO8601')
years=df_tracks.dates.dt.year
```

```python
sns.displot(years,discrete=True,aspect=2,kind="hist").set(title="Number of songs per year")
```

```python
total_dr=df_tracks.duration
fig_dims=(18,7)
fig,ax=plt.subplots(figsize=fig_dims)
fig=sns.barplot(x=years,y=total_dr,ax=ax,errwidth=False).set(title="YEar vs Duration")
plt.xticks(rotation=90)
```

```python
total_dr=df_tracks.duration
sns.set_style(style="whitegrid")
fig_dims=(10,5)
fig,ax =plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=years,y=total_dr,ax=ax).set(title="year vs duration")
plt.xticks(rotation=60)
```
